### Step 6. For each occupation, calculate the minimum and maximum ages

```python
user.groupby(by=['occupation',])[['age']].agg(['min','max']).reset_index()
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
user.groupby(by=['occupation','gender'])[['age']].mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
total_count=user.groupby(by=['occupation'])[['gender']].count()
ind_count=user.groupby(by=['occupation','gender'])[['gender']].count()
percentage=(ind_count/total_count*100).round(2)
```
